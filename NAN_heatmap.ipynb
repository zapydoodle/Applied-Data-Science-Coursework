```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from matplotlib.patches import Patch
```

```python
df = pd.read_csv('clean_dataset.csv')
df.head()
```

```python
# replace NA with NAN
missing_values_matrix = df.replace('NA', np.nan)
# change sex values to integers
missing_values_matrix['sex'] = missing_values_matrix['sex'].replace('Female', 1)
missing_values_matrix['sex'] = missing_values_matrix['sex'].replace('Male', 2)
# get boolean values
missing_values_matrix = missing_values_matrix.isna()
missing_values_matrix.head()

```

```python
missing_values_matrix.info()
```

```python
# Set up the plot size and style
plt.figure(figsize=(26, 8))
sns.set(style='white')

# Create the heatmap
sns.heatmap(missing_values_matrix, cmap='viridis', cbar=False, yticklabels=False)

# Customize the plot
plt.title('Missing Values Heatmap', fontsize=18)
plt.xlabel('Columns', fontsize=14)
plt.ylabel('Rows', fontsize=14)

# Create custom legend
legend_elements = [Patch(facecolor='purple', edgecolor='purple', label='Non-missing Value'),
                   Patch(facecolor='yellow', edgecolor='yellow', label='NA')]

plt.legend(handles=legend_elements, loc='upper center', bbox_to_anchor=(0.8, 1.11), ncol=3, fontsize=12, title='Value Status', title_fontsize=14)

# Show the plot
plt.show()
```
