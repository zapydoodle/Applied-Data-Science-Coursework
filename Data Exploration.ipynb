# What is the data and where does it come from?

* The data collected for this dataset was collected as part of the Avon Longitudinal Study of Parents and Children (ALSPAC), the data spans the childs life from pregnancy to age 18
* The data includes many features such as the persons screen time, their mothers level of depression, and the amount of exercise they do. 
* There are also several label type variables i.e. depression and anxiety scores. 
* The dataset is not complete by any means, there are a lot of missing entries

# Missing data

```python
import pandas as pd
import matplotlib.pyplot as plt

def count_nans(csv_filename, unused_columns):
    """
    Count the amount of NaNs in each column of a dataframe 
    """
    #Â Import in csv
    data = pd.read_csv(csv_filename)

    # Make dictionary from feature names
    NA_frequencies = dict.fromkeys(data.columns, None)

    # Remove columns not needed
    for column in unused_columns:
        NA_frequencies.pop(column)

    #Â Loop through keys
    for column in NA_frequencies:
        # Count the number of NAs in column
        NA_frequency = data[column].isnull().sum()
        #Â Add frequency to dict
        NA_frequencies[column] = NA_frequency

    # Sort dict in ascending order of frequencies
    NA_frequencies_sorted = {k: v for k, v in sorted(NA_frequencies.items(), key=lambda item: item[1])}

    datapoints = len(data)

    return NA_frequencies_sorted, datapoints

def plot_frequency(NA_frequencies, datapoints):
    # Make nice figure
    fig, ax = plt.subplots(figsize=(18,30))

    # Plot histogram with each frequency
    bar = plt.barh(*zip(*NA_frequencies.items()), height=0.85, align='center')

    # Amount of datapoints for csv:
    row_num = datapoints

    # Add counts above the two bar graphs
    for rect in bar:
        height = (rect.get_width()/row_num)*100
        plt.text(rect.get_width()+100, rect.get_y()+0.15, f'{height:.0f} %')

    plt.show()

# List of column headings where we're not interested in NaNs (will change depending on csv)
not_needed = ['Unnamed: 0', 'X', 'flag']

NA_frequencies, datapoints = count_nans("maps-synthetic-data-v1.1.csv", not_needed)

plot_frequency(NA_frequencies, datapoints)
```
